```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.linspace(-1,-5,100)
t=np.linspace(1,10,100)
v=[a[i]*ti for i, ti in enumerate(t)]

plt.plot(t,v)
plt.xlabel("time")
plt.ylabel("speed")
plt.title("Decceleration at an increasing rate")
plt.show()
```

```python
a=np.linspace(0,60,100)
t=np.linspace(1,10,100)
v=[a[i]*ti + 20 for i, ti in enumerate(t)]

plt.plot(t,v)
plt.xlabel("time")
plt.ylabel("speed")
plt.title("Increasing acceleration")
plt.show()
```
